# Linear Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
linear_model_sk = LinearRegression()
```

```python
linear_model_sk.fit(X=train[["X1", "X2"]],
                    y=train["y"])
```

```python
print(linear_model_sk.coef_)
print(linear_model_sk.intercept_)
```

```python
train["y_preds"] = linear_model_sk.predict(train[["X1", "X2"]])
print(f'Error Absoluto Promedio {round(mean_absolute_error(train["y"], train["y_preds"]), 2)}')
```

```python
test["y_preds"] = linear_model_sk.predict(test[["X1", "X2"]])
print(f'Error Absoluto Promedio {round(mean_absolute_error(test["y"], test["y_preds"]), 2)}')
```

# Dense Neural Net

```python
from tensorflow import keras
```

```python
model = keras.models.Sequential(
    [keras.layers.Dense(5, activation="sigmoid", input_shape=(2,)), 
     #keras.layers.Dense(3, activation="relu"), 
     keras.layers.Dense(1)])

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.1), loss="mae")

history = model.fit(x=train[["X1", "X2"]], y=train["y"], batch_size=10, epochs=500, shuffle=True)
```

```python
print(f'Error Absoluto Promedio {round(mean_absolute_error(train["y"], model.predict(train[["X1", "X2"]])), 2)}')
print(f'Error Absoluto Promedio {round(mean_absolute_error(test["y"], model.predict(test[["X1", "X2"]])), 2)}')
```

```python
plot = px.scatter_3d(data_frame=train,
              x="X1",
              y="X2",
              z="y",
              color=model.predict(train[["X1", "X2"]]).ravel(), 
              #color="y",
              color_continuous_scale="viridis",
              height=700,
              width=600,
              template="plotly_white")

#plot.add_trace(go.Surface(x=train["X1"].to_numpy(), y=train["X2"].to_numpy(), z=train["y_preds"].to_numpy()))

plot.show()
```
